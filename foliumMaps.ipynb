```python
import folium
import pandas as pd
```

```python
remove = False
```

```python
southSide = ['Doilier Street, Burgh Quay','Aston Quay', 'Westmoreland Street East',
            'Westmoreland Street West', 'College Green, Bank Of Ireland', 
            'Grafton Street @ CompuB', 'College Green @ Church Lane',
            'College Green - Dame St Side', 'Dame Street', 
             'Grafton Street / Nassau Street / Suffolk Street',
            'Grafton Street','Dawson Street Replacement','Dawson Street',
            'Grove Road Totem']
```

```python
pedestrianCounterLocations = pd.read_csv('dcc-footfall-counter-locations-14082020.csv')
if remove:
    pedestrianCounterLocations = pedestrianCounterLocations[~pedestrianCounterLocations['Counter Locations'].isin(southSide)]
pedestrianCounterLocations['Counter Locations'] = pedestrianCounterLocations['Counter Locations'] + ' Pedestrian Counter'
cycleCounterLocations = pd.read_csv('dcc_nta_cycle-counter-locations-14082020.csv')
cycleCounterLocations.rename(columns={'Bike Counter Locations':'Counter Locations'}, inplace=True)
if remove:
    cycleCounterLocations = cycleCounterLocations[~cycleCounterLocations['Counter Locations'].isin(southSide)]
cycleCounterLocations['Counter Locations'] = cycleCounterLocations['Counter Locations'] + ' Cycle Counter'

totalData = pd.concat([pedestrianCounterLocations, cycleCounterLocations], ignore_index=True)
totalData
```

```python
m = folium.Map(location=[53.349020, -6.260050], zoom_start=13)

for index, row in totalData.iterrows():
    if row['User Type'] == 'Pedestrians':
        markerColor = 'green'
    else:
        markerColor = 'blue'
    folium.Marker(
        [row['Latitude'], row['Longitude']],
    popup=row['Counter Locations'],
    icon=folium.Icon(color=markerColor)).add_to(m)
m
```

```python
folium.Marker(
        [row['Latitude'], row['Longitude']],
    popup=row['Counter Locations'],
    icon=folium.Icon(color=markerColor, icon_color='white', icon='male', prefix='fa')).add_to(m)
m
```
